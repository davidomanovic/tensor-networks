```python
import matplotlib.pyplot as plt
import numpy as np
import tikzplotlib

data = np.loadtxt("../data/dmrg/sweeps/N2_6-31g_eq.csv",delimiter=',',skiprows=1)

chi = data[:,0]
E = data[:,1]

e_true = E[-1]
Es = E[:-1]
chis = chi[:-1]

fig, (ax1,ax2) = plt.subplots(1,2, figsize=(16,5))

ax1.plot(chi, E, ':o', c="b")
ax1.axhline(E[-1],c='gray',ls=':')
ax1.set_xscale('log', base=2)
ax1.set_xlabel(r"Max DMRG bond dimension $\chi$")
ax1.set_ylabel(r"$E_{eq}$ [Ha]")

ax2.semilogy(chis, np.abs(Es-e_true), ':x', c='k', label="Error from converged solution")
ax2.axhline(1E-3, ls=':', c='gray', label="Chemical accuracy")
ax2.set_ylabel(r"Absolute error $|\varepsilon|$ [Ha]")
ax2.set_xlabel(r"Max DMRG bond dimension $\chi$")
ax2.set_xscale('log', base=2)

tikzplotlib.save( "figures/N2_6-31g.tex", axis_width="0.5\\linewidth", axis_height="0.4\\linewidth", textsize=6, extra_axis_parameters=[ "tick label style={font=\\scriptsize}", "label style={font=\\footnotesize}", "title style={font=\\footnotesize}", "legend style={font=\\scriptsize}", ], extra_groupstyle_parameters=[ "horizontal sep=1.75cm" ] )

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import tikzplotlib

ds = [("../data/dmrg/dmrgci/n2_ccpvdz(10e,26o)_bd=1024.csv", r"DMRG $\chi=1024$", "k", "",2,1)]
plt.figure(figsize=(8,6))
dataz = np.loadtxt("../data/classical/n2_HF_ccpvdz.csv", delimiter=',', skiprows=1)
for (data, label, c, ls, lw, ms) in ds:
    dataz = np.loadtxt(data, delimiter=',', skiprows=1)
    R, E = dataz[:,0], dataz[:,1]
    plt.plot(R, E, "-", c=c, lw=1)
    plt.plot(R, E, f"{ls}", label=label, c=c, lw=lw, markersize=ms, markerfacecolor=c, fillstyle="none")
    plt.xlabel(r"(N=N) bond distance $R$ []")
    plt.ylabel(r"$E$ [Ha]")

data = np.loadtxt("../data/mps/cc-pvdz/n2_bd=512_stride=4.csv",delimiter=',',skiprows=1)
R = data[:,0]
E = data[:,1]
plt.plot(R,E, '-',c='magenta')
plt.plot(R,E, 'x',c='magenta', lw=5, label=r"MPS LUCJ, $\chi=512$")

data = np.loadtxt("../data/classical/n2_cc-pvdz.csv",delimiter=',',skiprows=1)
R = data[:,0]
E = data[:,2]
plt.plot(R,E, '-',c='orange')
plt.plot(R,E, 'x',c='orange', lw=5, label=r"CCSD")

plt.legend(framealpha=0, loc="upper center", bbox_to_anchor=(0.5,1.2), ncol=2, fontsize=10)
tikzplotlib.save("figures/N2_dmrgci_ccpvdz.tex",
                 axis_width="0.8\\linewidth", 
                 axis_height="0.5\\linewidth",
                 extra_axis_parameters=[r"font=\small",
                                        r"every axis plot/.append style={mark options={solid}}",
                                        r"legend image post style={mark options={solid}, line width=1.5pt}",])
plt.show()
```
