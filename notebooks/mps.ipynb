```python
import matplotlib.pyplot as plt
import numpy as np
import tikzplotlib


bds = [(62,'o', 'magenta'),(63,'d', 'green'), (64,'o', 'red')]

ground_truth = np.loadtxt("../data/classical/n2_sto-6g.csv",delimiter=',',skiprows=1)
R = ground_truth[:,0]
E_lucj = ground_truth[:,5]

fig, axs = plt.subplots(1,2, figsize=(10,4))

axs[0].plot(R,E_lucj, c='k', label="LuCJ")
for (bd, mk, c) in bds:
    data = np.loadtxt(f"../data/mps/sto-6g/n2_bd={bd}.csv",delimiter=',',skiprows=1)
    R = data[:,0]; E=data[:,1]
    axs[0].plot(R,E, mk,c=c, label=rf"$\chi=${bd} $\,$", markersize=2)
axs[0].set_xlabel(r"Bond distance []"); axs[0].set_ylabel(r"E [Ha]")
axs[0].set_ylim(-108.65,-107.5)
axs[0].legend(ncol=4, loc="upper left", framealpha=1)

for (bd, mk, c) in bds:
    data = np.loadtxt(f"../data/mps/sto-6g/n2_bd={bd}.csv",delimiter=',',skiprows=1)
    R = data[:,0]; E=data[:,1]
    axs[1].semilogy(R, np.abs(E - E_lucj), mk, c=c, label=rf"$\chi=${bd}", markersize=2,ls=':')
axs[1].set_xlabel(r"Bond distance []"); axs[1].set_ylabel(r"$|\Delta E|$ [Ha]")
axs[1].axhline(1E-3, c='gray', ls="--")

plt.tight_layout()


tikzplotlib.save( "figures/N2_sto-6g_MPS.tex", axis_width="0.5\\linewidth", axis_height="0.4\\linewidth", textsize=4, 
                 extra_axis_parameters=[ "tick label style={font=\\scriptsize}", 
                                        "label style={font=\\footnotesize}", 
                                        "title style={font=\\footnotesize}", 
                                        "legend style={font=\\scriptsize}"], 
                                        extra_groupstyle_parameters=[ "horizontal sep=1.75cm" ] )

plt.show()
```
