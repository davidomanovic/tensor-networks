# Analysis of N2 and H2 sweeps over different ans√§tze

```python
import tikzplotlib
import matplotlib.pyplot as plt
import numpy as np
from cycler import cycler

plt.rcParams.update({"lines.markersize":3,
                     "xtick.major.size":3,
                     "xtick.minor.size":3,
                     "ytick.major.size":3,
                     "ytick.minor.size":3,
                     "xtick.direction":"in",
                     "ytick.direction":"in"})

plt.rcParams["axes.prop_cycle"] = cycler(color=["#ff0000", "#ff7700", "#00ff00", "#00ffd5", "#1500ff", "#9000ff", "#ff00b3", "#646464", "#ffff00", "#00ffb3"])
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Load data
d = pd.read_csv("../data/classical/h2_sto-3g.csv")
R = d["R_AA"].to_numpy()
E_FCI = d["E_FCI"].to_numpy()

series = {
    "CCSD": d["E_CCSD"].to_numpy(),
    "uCJ": d["E_uCJ"].to_numpy(),
    "2-uCJ": d["E_2uCJ"].to_numpy(),
    "LuCJ": d["E_L-1uCJ"].to_numpy(),
    "2-LuCJ": d["E_L-2uCJ"].to_numpy(),
    "1-LuCJ(opt)": d["E_L-1uCJ_opt"].to_numpy(),
    "2-LuCJ(opt)": d["E_L-2uCJ_opt"].to_numpy(),
}

# Absolute errors vs FCI
err = {k: np.abs(v - E_FCI) for k, v in series.items()}

# Marker styles you asked for
styles = {
    "CCSD": {"ls": "-", "marker": ""},
    "uCJ": {"ls": "-", "marker": "x"},
    "2-uCJ": {"ls": "-", "marker": "x"},
    "LuCJ": {"ls": "-", "marker": "^"},
    "2-LuCJ": {"ls": "-", "marker": "^"},
    "1-LuCJ(opt)": {"ls": "-", "marker": "s"},
    "2-LuCJ(opt)": {"ls": "-", "marker": "s"},
}

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(20, 10), sharex=True)

ax1.plot(R, E_FCI, label="FCI", c="k", lw=3)
for lab, E in series.items():
    ax1.plot(R, E, label=lab, **styles.get(lab, {}), markersize=2)

ax1.set_ylabel("E [Ha]")
ax1.set_xlabel("(H-H) intermolecular distance [√Ö]")
ax1.legend(ncol=4, loc="upper center", bbox_to_anchor=(1.1, 1.25))
ax1.set_xlim(R[0], R[-1])

for lab, e in err.items():
    ax2.semilogy(R, np.maximum(e, 1e-16), label=lab, **styles.get(lab, {}), markersize=2)

ax2.set_ylabel("Absolute error")
ax2.set_xlabel("(H-H) intermolecular distance [√Ö]")
ax2.set_xlim(R[0], R[-1])

plt.tight_layout()
tikzplotlib.save( "figures/H2_sto-3g.tex", axis_width="0.5\\linewidth", axis_height="0.4\\linewidth", textsize=6, extra_axis_parameters=[ "tick label style={font=\\scriptsize}", "label style={font=\\footnotesize}", "title style={font=\\footnotesize}", "legend style={font=\\scriptsize}", ], extra_groupstyle_parameters=[ "horizontal sep=1.75cm" ] )
```
